## Paso 1: Generar lista de notas

```python
import random
notas = random.sample(range(0, 21), 15)  # 15 notas entre 0 y 20 sin repetidos
print("üìã Notas originales:", notas)
```

```python
# Paso 2: Bubble Sort con visualizaci√≥n

import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

def bubble_sort_visual(notas):
    pasos = [notas.copy()]
    n = len(notas)
    for i in range(n):
        for j in range(0, n - i - 1):
            if notas[j] > notas[j + 1]:
                notas[j], notas[j + 1] = notas[j + 1], notas[j]
                pasos.append(notas.copy())
    return pasos

# Ejecutar y visualizar los pasos
pasos_bubble = bubble_sort_visual(notas.copy())
for paso, estado in enumerate(pasos_bubble):
    clear_output(wait=True)
    plt.bar(range(len(estado)), estado, color='skyblue')
    plt.title(f"Bubble Sort - Paso {paso+1}")
    plt.xlabel("√çndice")
    plt.ylabel("Nota")
    plt.ylim(0, 22)
    plt.pause(0.3)
plt.show()
```

## Paso 2: Visualizar Bubble Sort (puedes guiarte de `4_2_Visualizacion_BubbleSort.ipynb`)

```python
# Paso 3: Simulaci√≥n visual de sorted()

notas_sorted = sorted(notas.copy())  # resultado final
simulacion = []
temporal = notas.copy()

# Simulamos que se construye la lista sorted paso a paso
for i in range(len(notas_sorted)):
    if temporal[i] != notas_sorted[i]:
        temporal[i] = notas_sorted[i]
    simulacion.append(temporal.copy())

for k, estado in enumerate(simulacion):
    clear_output(wait=True)
    plt.bar(range(len(estado)), estado, color='lightgreen')
    plt.title(f"Simulaci√≥n sorted() - Paso {k+1}")
    plt.xlabel("√çndice")
    plt.ylabel("Nota")
    plt.ylim(0, 22)
    plt.pause(0.3)
plt.show()
```

## Paso 4: Insertar una nueva calificaci√≥n con bisect (ver `4_1_Clase_Complementaria_Bisect_Sorted.ipynb`)

## Paso 5: Buscar una calificaci√≥n usando bisect

```python
# ‚úÖ Aseguramos que la lista est√© ordenada antes de usar bisect
notas.sort()

print("üìã Lista ordenada:", notas)

# Valor a buscar
nota_buscar = 10

# Usamos bisect_left para obtener la posici√≥n donde estar√≠a esa nota
posicion = bisect.bisect_left(notas, nota_buscar)

# Verificamos si realmente est√° en esa posici√≥n
if posicion < len(notas) and notas[posicion] == nota_buscar:
    print(f"üîç La nota {nota_buscar} se encuentra en la posici√≥n {posicion}")
else:
    print(f"‚ùå La nota {nota_buscar} no est√° en la lista.")
```
