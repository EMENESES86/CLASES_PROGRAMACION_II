## Paso 1: Generar lista de notas

```python
import random
notas = random.sample(range(0, 21), 15)  # 15 notas entre 0 y 20 sin repetidos
print("📋 Notas originales:", notas)
```

```python
# Paso 2: Bubble Sort con visualización

import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

def bubble_sort_visual(notas):
    pasos = [notas.copy()]
    n = len(notas)
    for i in range(n):
        for j in range(0, n - i - 1):
            if notas[j] > notas[j + 1]:
                notas[j], notas[j + 1] = notas[j + 1], notas[j]
                pasos.append(notas.copy())
    return pasos

# Ejecutar y visualizar los pasos
pasos_bubble = bubble_sort_visual(notas.copy())
for paso, estado in enumerate(pasos_bubble):
    clear_output(wait=True)
    plt.bar(range(len(estado)), estado, color='skyblue')
    plt.title(f"Bubble Sort - Paso {paso+1}")
    plt.xlabel("Índice")
    plt.ylabel("Nota")
    plt.ylim(0, 22)
    plt.pause(0.3)
plt.show()
```

## Paso 2: Visualizar Bubble Sort (puedes guiarte de `4_2_Visualizacion_BubbleSort.ipynb`)

```python
# Paso 3: Simulación visual de sorted()

notas_sorted = sorted(notas.copy())  # resultado final
simulacion = []
temporal = notas.copy()

# Simulamos que se construye la lista sorted paso a paso
for i in range(len(notas_sorted)):
    if temporal[i] != notas_sorted[i]:
        temporal[i] = notas_sorted[i]
    simulacion.append(temporal.copy())

for k, estado in enumerate(simulacion):
    clear_output(wait=True)
    plt.bar(range(len(estado)), estado, color='lightgreen')
    plt.title(f"Simulación sorted() - Paso {k+1}")
    plt.xlabel("Índice")
    plt.ylabel("Nota")
    plt.ylim(0, 22)
    plt.pause(0.3)
plt.show()
```

## Paso 5: Buscar una calificación usando bisect

```python
# ✅ Aseguramos que la lista esté ordenada antes de usar bisect
notas.sort()

print("📋 Lista ordenada:", notas)

# Valor a buscar
nota_buscar = 10

# Usamos bisect_left para obtener la posición donde estaría esa nota
posicion = bisect.bisect_left(notas, nota_buscar)

# Verificamos si realmente está en esa posición
if posicion < len(notas) and notas[posicion] == nota_buscar:
    print(f"🔍 La nota {nota_buscar} se encuentra en la posición {posicion}")
else:
    print(f"❌ La nota {nota_buscar} no está en la lista.")
```
